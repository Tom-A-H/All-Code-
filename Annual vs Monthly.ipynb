```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 
import cartopy.crs as ccrs
import cmocean
from OceanDataStore import OceanDataCatalog
```

```python
catalog = OceanDataCatalog(catalog_name="noc-model-stac")
catalog.search(collection='noc-npd-era5', standard_name='sea_surface_temperature')
catalog.available_items
ds_annual = catalog.open_dataset(id=catalog.Items[2].id,
                          start_datetime='1990-01',
                          end_datetime='2024-12')
ds_monthly = catalog.open_dataset(id=catalog.Items[4].id,
                          start_datetime='1990-01',
                          end_datetime='2024-12')
```

```python
rp_annual = (ds_annual['tos_con'].sel(x = 600, y = 900)).values
rp_monthly = (ds_monthly['tos_con'].sel(x = 600, y = 900)).values
```

```python
rp_monthly = rp_monthly.values
```

```python
rp_monthly
```

```python
calculated_annual = []
count = 0
total = 0
for i in range (0, len(rp_monthly)):
    total = total + rp_monthly[i]
    count = count + 1
    if count == 73:
        calculated_annual.append(np.round((total/73), 6))
        total = 0
        count = 0

print(rp_annual)
print(calculated_annual) 
```
