```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 
import cartopy.crs as ccrs
import cmocean
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
from OceanDataStore import OceanDataCatalog
from matplotlib.patches import Rectangle
from nemo_cookbook import NEMODataTree  
```

```python
# Open model and config files
catalog = OceanDataCatalog(catalog_name="noc-model-stac")
catalog.search(collection='noc-npd-era5', standard_name='sea_surface_temperature')
ds1 = catalog.open_dataset(id=catalog.Items[2].id,
                          start_datetime='1990-01',
                          end_datetime='2024-12')

catalog.search(collection='noc-npd-era5', item_name='domain_cfg')
config = catalog.open_dataset(id=catalog.Items[1].id)

# Merge into a data tree
datasets = {"parent": {"domain": config, "gridT": ds1}}
dt = NEMODataTree.from_datasets(datasets = datasets)

# Clip to North Atlantic 
bbox = (-85.0, 0.0, 0.0, 80.0)
dt_clipped = dt.clip_grid(grid="/gridT", bbox=bbox)

Add lat and lon as co-ordinates
dt = dt_clipped.add_geoindex(grid="/gridT")

# Convert to dataset
ds = (dt_clipped['/gridT']).dataset

```

```python
data = ds['sos_abs'].compute()
```

```python
ds
```

```python
## Plotting Map of salinity trends 

ny, nx = data.sizes['j'], data.sizes['i']
trend_data = np.full((ny, nx), np.nan, dtype=np.float32)

for y_idx in range (ny):
    for x_idx in range (nx):
        point = data.isel(j=y_idx, i=x_idx)
        try:
            z = np.polyfit(point['time_counter'].dt.year, point.values, 1)
            trend_data[y_idx, x_idx] = z[0]
        except:
            trend_data[y_idx, x_idx] = np.nan

trend_da = xr.DataArray(data = trend_data, dims=["j", "i"], 
        coords={ "j": data['j'],
        "i": data['i'],
        "lat": (("j", "i"), data['gphit'].values),
        "lon": (("j", "i"), data['glamt'].values)}, name="trend",
        attrs={"description": "Linear trend over time of SSS"})

fig, ax = plt.subplots(subplot_kw={'projection': ccrs.PlateCarree()}, figsize=(10, 6))
im = ax.pcolormesh(trend_da['lon'], trend_da['lat'], trend_da, cmap='RdBu_r', transform=ccrs.PlateCarree(), vmin= -0.1, vmax=0.1)  
ax.set_extent([-85.0, 0.0, 0.0, 80.0], crs=ccrs.PlateCarree())

cl = ax.coastlines()
gl = ax.gridlines(draw_labels=True, linewidth=0.5, linestyle='--')
box = Rectangle((-45, 40), 25, 25, linewidth=2, edgecolor='black', 
                facecolor='none', linestyle='--', transform=ccrs.PlateCarree())
ax.add_patch(box)
ax.plot(-36, 53, 'ko', markersize=8, transform=ccrs.PlateCarree())
plt.colorbar(im, ax=ax)
plt.tight_layout()
title = ax.set_title('1990-2024 trends in sea surface salinity')  
plt.savefig('Salinity Trend Map.png') 
```
